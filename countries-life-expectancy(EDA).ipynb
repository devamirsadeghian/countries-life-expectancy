### EDA

```python
# میانگین تاریخ تولد افراد دیتاست چقدر است
df['Year'].mean()
```

```python
# از هر کشور چه تعداد افرادی وجود دارد
df['Entity'].value_counts()
```

```python
# فقط نام کشور های مختلف را پیدا کن
df['Entity'].unique()
```

```python
#افراد مربوط به چند کشور مختلف هستند
df['Entity'].value_counts().count()
```

```python
# دیتاست را اول بر اساس سال تولد افراد از کوچک به بزرگ و دوما بر اساس مقدار امید به زندگی افراد از کوچک به بزرگ مرتب کن
df.sort_values(['Year','Life expectancy'])
```

```python
# میزان همبستگی به چه شکل است
df.corr()
# با توجه به میزان همبستگی بالا ، هر چقدر فرد جوان تر باشد میزان امید به زندگی بالا تر است
# ستون میزان امید به زنگی و سال تولد افراد همبستگی خیلی بالایی دارد
```

```python
sns.pairplot(df , height=6 , aspect=1)
```

```python
sns.jointplot( data=df , x='Year' , y='Life expectancy' , kind='hex' , color='red')
```

```python
plt.figure(figsize=(15,10))
sns.histplot( data=df , x='Year' , kde=True  , bins=200)
plt.title('تعداد متولدین هر سال')
```

```python
plt.figure(figsize=(16,8))
sns.barplot( data=df , x='Entity' , y='Life expectancy' )
```

```python
plt.figure(figsize=(14,8))
sns.boxplot( data=df , x='Entity' , y='Life expectancy' ,palette='rainbow_r')
```
